## Line Chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 10, 10)
y = np.sin(x)
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.xlabel("Time")
plt.ylabel("Some function of time")
plt.title("Cool")
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

## Scatter Plot

```python
import pandas as pd
```

```python
A = pd.read_csv("data_1d.csv", header=None)
```

```python
x = A[0]
y = A[1]
```

```python
plt.scatter(A[0], A[1])
```

## Histogram

```python
x = A[0]
plt.hist(x)
```
